# Gráficos iniciais

```python
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd 
import seaborn as sns

```

```python
d = pd.read_csv('testando.csv')

d
```

```python
sns.boxplot(x='class',y='fft',hue='class',data=d,showfliers=False)
```

```python
fig = plt.figure(figsize=(16,8))
plt.plot(d[d['class']=='yes'].loc[:,'fft'].to_numpy(),'.')
plt.plot(d[d['class']=='no'].loc[:,'fft'].to_numpy(),'.')

plt.show()
```
