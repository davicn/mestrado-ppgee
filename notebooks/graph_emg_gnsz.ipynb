# Gráficos iniciais

```python
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd 
import seaborn as sns

```

```python
d = pd.read_csv('testando.csv')

d
```

```python
import matplotlib.gridspec as gridspec

# fig = plt.figure(figsize=(10,10))
fig, ax = plt.subplots(2,2, figsize=(8,8), sharey=True, dpi=120)

# gs = gridspec.GridSpec(2,2)

sns.boxplot(x='class',y='fft',hue='class',data=d,showfliers=False,ax=ax[0,0])
sns.boxplot(x='class',y='var',hue='class',data=d,showfliers=False,ax=ax[0,1])
sns.boxplot(x='class',y='kur',hue='class',data=d,showfliers=False,ax=ax[1,0])
sns.boxplot(x='class',y='skew',hue='class',data=d,showfliers=False,ax=ax[1,1])
```

```python
axs = plt.subplot2grid(2,2)
plt.plot(d[d['class']=='yes'].loc[:,'fft'].to_numpy(),'.')
plt.plot(d[d['class']=='no'].loc[:,'fft'].to_numpy(),'.')
plt.semilogy()
plt.show()
```
